- **Get Trained model**

```python
# importing dependancies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
%matplotlib inline
```

# **Data Collection**

```python
sonar_data = pd.read_csv('/content/sonar data.csv', header = None)
sonar_data.head()
```

# **Data Preprocessing**

```python
sonar_data.info()
```

```python
sonar_data.describe()
```

```python
# checking the counts of rocks and mines
sonar_data[60].value_counts()
```

```python
# grouping 60 and getting mean - there is a diffrence in mean of both r and m
sonar_data.groupby(60).mean()
```

```python
# separating columns
X = sonar_data.iloc[:, :-1] # Independent Column
Y = sonar_data.iloc[:, -1] # Target/Dependent column
```

```python
print(X, Y)
```

# **Splitting data**

```python
# Splitting for Training and Testing data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.1, stratify = Y, random_state = 1)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

# **Model training**

```python
# Model training - Logestic Regression
model = LogisticRegression()
model
```

```python
# Training model
model.fit(X_train, y_train)
```

# **Model Evaluation**

```python
# Accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
print(f'Accuracy Score: {training_data_accuracy}')
```

```python
# Accuracy score on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
print(f'Accuracy Score: {test_data_accuracy}')
```
